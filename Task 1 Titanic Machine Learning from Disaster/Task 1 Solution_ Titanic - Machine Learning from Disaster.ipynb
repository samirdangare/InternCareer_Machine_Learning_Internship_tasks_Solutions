### Data Cleaning

```python
df_train['Age_Cat'] = df_train['Age'].apply(lambda Age:'Child' if Age<18 else 'Non_Child')
df_train['Age_Cat']
```

```python
v=df_train['Age'].mean()
df_train['Age'].fillna(v,inplace=True)
df_test['Age'].fillna(v,inplace=True)


v=df_train['Fare'].mean()
df_train['Fare'].fillna(v,inplace=True)
df_test['Fare'].fillna(v, inplace=True)
```

```python
y_train=df_train['Survived']
X_train=df_train[['Pclass','SibSp','Parch','Age','Fare']]
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
# Make predictions
```

```python
X_test=df_test[['Pclass','SibSp','Parch','Age','Fare']]
yp=model.predict(X_test)
```
